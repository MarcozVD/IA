```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset simulado
np.random.seed(42)
data = pd.DataFrame({
    "Edad": np.round(np.random.normal(30, 10, 100).tolist() + [100, 120]),  # Valores atípicos
    "Ingresos": np.round(np.random.normal(50000, 15000, 100).tolist() + [200000, 250000])
})
data
```

```python
from sklearn.preprocessing import MinMaxScaler #escalamos los datos entre 1 como maximo y 0 como el minimo
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(data)
X_scaled=pd.DataFrame(X_scaled,columns=['Edad','Ingresos'])
X_scaled
```
